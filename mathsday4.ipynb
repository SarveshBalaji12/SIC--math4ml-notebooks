```python
def min_max(data):
    min_val=min(data)
    max_val=max(data)
    return [(x-min_val)/(max_val-min_val) for x in data]

data=[50,20,30,10,90,60]
scaled_data=min_max(data)
print(scaled_data)
```

```python
import numpy as np
def logf(x):
    return np.log(x+1)

def stan(x,mean,std):
    return (x-mean)/std

def composite(x,mean,std):
    return stan(logf(x),mean,std)

data=[10,50,100,200,500]
mean=np.mean([logf(x) for x in data])
std=np.std([logf(x) for x in data] )

tran_data=[composite(x,mean,std) for x in data]
print(tran_data)
```

```python
def z_score(x,mean,std):
    return (x-mean)/std

def inv(z,mean,std):
    return z*std+mean

data=[100,200,300,400,500]
mean=sum(data)/len(data)
std=(sum((x-mean)**2 for x in data)/len(data))**0.5

nor_data=[z_score(x,mean,std) for x in data]
ori_data=[inv(z,mean,std) for z in nor_data]

print(nor_data)
print(ori_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-2,4,100)

y1=3**x
y2=3**(x-1)+2

plt.plot(x,y1,label=r"$y=3^x$")
plt.plot(x,y2,label=r"$y=3^{(x-1)}+2$")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
