```python
# Import Libraries
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load Dataset
iris = datasets.load_iris()
X, y = iris.data, iris.target

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Polynomial SVM Model
model = SVC(kernel='rbf', degree=3, C=1.0)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
print(f"Accuracy on Iris Dataset: {accuracy_score(y_test, y_pred):.4f}")
iris
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load Dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
column_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
                'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']

df = pd.read_csv(url, names=column_names)
X, y = df.iloc[:, :-1].values, df.iloc[:, -1].values

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Polynomial SVM Model
model = SVC(kernel='linear', C=0.01)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
print(f"Accuracy on Pima Diabetes Dataset: {accuracy_score(y_test, y_pred):.4f}")
df
```

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate Spiral Dataset
def create_spiral_dataset(n_samples, noise=0.2):
    n = n_samples // 2
    t = np.linspace(0, 2 * np.pi, n)
    r = np.linspace(0, 1, n)
    X1 = np.array([r * np.sin(t), r * np.cos(t)]).T + noise * np.random.randn(n, 2)
    X2 = np.array([r * np.sin(t + np.pi), r * np.cos(t + np.pi)]).T + noise * np.random.randn(n, 2)
    X = np.vstack([X1, X2])
    y = np.hstack([np.zeros(n), np.ones(n)])
    return X, y

# Create Spiral Dataset
X, y = create_spiral_dataset(400)

# Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Polynomial SVM Model
model = SVC(kernel='rbf',gamma=1 , C=1.0)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
print(f"Accuracy on Spiral Dataset: {accuracy_score(y_test, y_pred):.4f}")
```
