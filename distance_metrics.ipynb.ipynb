```python
import numpy as np
p=np.array([2,3,4])
q=np.array([1,0,-1])
eucli_dist=np.linalg.norm(p-q)
man_dist=np.sum(np.abs(p-q))
cheby_dist=np.max(np.abs(p-q))
print(eucli_dist)
print(man_dist)
print(cheby_dist)
```

```python
def f(x):
    return 2*x+3

import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
y=f(x)
plt.plot(x,y,label="f(x)=2x+3",color='blue')
plt.grid(True)
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.legend()

plt.show()
```

```python
def f(x):
    return x**2-4*x+3

import matplotlib.pyplot as plt
x=np.linspace(-2,6,100)
y=f(x)
plt.plot(x,y,label="f(x)=x^2-4*x+3",color='blue')
plt.grid(True)
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.legend()

plt.show()
```

```python
def f(x):
    return np.exp(x)

import matplotlib.pyplot as plt
x=np.linspace(-2,2,100)
y=f(x)
plt.plot(x,y,label="f(x)=x^2-4*x+3",color='blue')
plt.grid(True)
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.legend()

plt.show()
```

```python
def f(x):
    return 1/(1+np.exp(-x))

import matplotlib.pyplot as plt
x=np.linspace(-4,4,100)
y=f(x)
plt.plot(x,y,label="f(x)=x^2-4*x+3",color='blue')
plt.grid(True)
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.legend()

plt.show()
```

```python
def f1(x):
    return np.sqrt(x+2)

def f(x):
    return x**3+1
print(f(f1(2)))
print(f1(f(2)))
print(f1(f1(2)))
print(f(f(2)))
```

```python
from sympy import symbols
x=symbols('x')

def f(x):
    return x**2+10

def g(x):
    return x+3

print(f(g(x)))
print(g(f(x)))
print(g(g(x)))
print(f(f(x)))
```
